```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
iris_data=pd.read_csv("iris.csv")
```

```python
iris_data.head()
```

```python
data=iris_data.drop("Id",axis=1)
```

```python
print("Data Shape",data.shape)
data.describe()
```

```python
print("We have 150 Data Points and 4 Features(Attributes) in IRIS Dataset")
```

```python
data.columns
```

```python
# Scatter plot between Sepal_Length and Sepal Width
sns.set()
sns.scatterplot(x='SepalLengthCm',y='SepalWidthCm',hue="Species",style='Species',data=data)
plt.show()
```

```python
sns.set()
sns.lineplot(x='SepalLengthCm',y='SepalWidthCm',hue="Species",data=data)
plt.show()
```

```python
sns.set()
sns.scatterplot(x='SepalLengthCm',y='PetalLengthCm',hue="Species",style='Species',data=data)
plt.show()
```

```python
sns.set()
sns.lineplot(x='SepalLengthCm',y='PetalLengthCm',hue="Species",data=data)
plt.show()
```

```python
sns.set()
sns.scatterplot(x='SepalWidthCm',y='PetalLengthCm',hue="Species",style='Species',data=data)
plt.show()
```

```python
sns.set()
sns.scatterplot(x='SepalWidthCm',y='PetalWidthCm',hue="Species",style='Species',data=data)
plt.show()
```

```python
sns.set()
sns.scatterplot(x='PetalLengthCm',y='PetalWidthCm',hue="Species",style='Species',data=data)
plt.show()
```

```python
sns.set(style="whitegrid")
sns.pairplot(data=data,hue="Species")
plt.savefig("PairPlot.jpeg",dpi=1080)
plt.show()
```

```python
#StripPlot
sns.set(style="whitegrid")
sns.stripplot(data=data)
plt.show()
```

```python
# SwarmPlot
sns.set(style="whitegrid")
sns.swarmplot(data=data,orient="v")
plt.show()
```

```python
sns.set(style="darkgrid")
sns.stripplot(x="Species",y="SepalLengthCm",data=data)
plt.plot()
```

```python
# Box Plot
sns.set_style("whitegrid")
sns.boxplot(data=data)
plt.show()
```

```python
# Box Plot
sns.set_style("whitegrid")
sns.boxplot(x="Species",y="PetalWidthCm",data=data)
plt.show()
```

```python
# ViolinPlot
sns.set(style="whitegrid")
sns.violinplot(x="Species",y="PetalWidthCm",data=data)
plt.show()
```

```python
#pointPlot
sns.set(style="whitegrid")
sns.pointplot(x="Species",y="PetalWidthCm",data=data)
plt.show()
```

```python
# barplot
sns.set(style="whitegrid")
sns.barplot(x="Species",y="PetalWidthCm",data=data)
plt.show()
```

```python
# Distributio Plot of the IRIS DataSet
sns.set()
sns.distplot(data[data['Species']=="Iris-setosa"].SepalLengthCm,bins=15,kde=False)
plt.show()
```

```python
# Distributio Plot of the IRIS DataSet
sns.set()
sns.distplot(data[data['Species']=="Iris-setosa"].SepalLengthCm,bins=15,hist=False)
plt.show()
```
